```python
# ============================
# Water Potability: MLP, TabNet, FT-Transformer
# - 5-fold cross-validation
# - SHAP explanations for all models (subset-based, GPU-aware)
# - Richer visualizations
# - Saves scaler and best model retrained on full data
# ============================

# ---------- Colab installs ----------
# Uncomment the next cell and run once if needed
!pip -q install pytorch-tabnet shap torch torchvision torchaudio --upgrade
!pip -q install scikit-learn matplotlib seaborn pandas numpy joblib

import os, time, math, random, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from functools import partial
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, f1_score, roc_auc_score, roc_curve,
    classification_report, confusion_matrix
)

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

from pytorch_tabnet.tab_model import TabNetClassifier
import shap

# ---------- Settings ----------
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", DEVICE)

DATA_PATH = "water_quality_potability.csv"  # change if needed or upload in Colab
BATCH_SIZE = 128
MLP_EPOCHS = 40
FTT_EPOCHS = 50
TABNET_MAX_EPOCHS = 100
LR = 1e-3
N_SPLITS = 5

ARTIFACT_DIR = "artifacts"
os.makedirs(ARTIFACT_DIR, exist_ok=True)

sns.set(style="whitegrid")

# ---------- Load ----------
df = pd.read_csv(DATA_PATH)
print("Dataset shape:", df.shape)
display(df.head())
assert "Potability" in df.columns, "Dataset must have 'Potability' column."

features = df.columns.tolist()
features.remove("Potability")
X_raw_all = df[features].copy()
y_all = df["Potability"].values

# ---------- Compact EDA ----------
print("\nTarget distribution:")
print(df["Potability"].value_counts())

plt.figure(figsize=(5,3))
sns.countplot(x="Potability", data=df)
plt.title("Class distribution (Potability)")
plt.show()

plt.figure(figsize=(9,7))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation heatmap")
plt.show()

top_feats = df[features].var().sort_values(ascending=False).index[:6].tolist()
for col in top_feats:
    plt.figure(figsize=(6,3.2))
    sns.kdeplot(data=df, x=col, hue="Potability", fill=True, common_norm=False)
    plt.title(f"Distribution of {col} by Potability")
    plt.tight_layout()
    plt.show()

# Pairplot on top 4 features (kept small to avoid long runtimes)
pp_feats = top_feats[:4]
sns.pairplot(df[pp_feats + ["Potability"]], hue="Potability", diag_kind="kde", corner=True)
plt.suptitle("Pairplot (top 4 features)", y=1.02)
plt.show()

# ---------- Utilities ----------
def plot_confusion(cm, title="Confusion matrix", normalize=False):
    plt.figure(figsize=(4.2,3.6))
    if normalize:
        cm_sum = cm.sum(axis=1, keepdims=True)
        cm = cm / (cm_sum + 1e-9)
        fmt = ".2f"
    else:
        fmt = "d"
    sns.heatmap(cm, annot=True, fmt=fmt, cmap="Blues")
    plt.title(title)
    plt.ylabel("Actual")
    plt.xlabel("Predicted")
    plt.show()

def evaluate_model(y_true, y_pred, y_prob=None, model_name="Model", plot_roc=False, ax=None):
    print(f"\n-- {model_name} Evaluation --")
    print(classification_report(y_true, y_pred, digits=4))
    cm = confusion_matrix(y_true, y_pred)
    plot_confusion(cm, title=f"{model_name} Confusion Matrix")
    plot_confusion(confusion_matrix(y_true, y_pred), title=f"{model_name} Confusion Matrix (Normalized)", normalize=True)

    auc = None
    if y_prob is not None:
        try:
            auc = roc_auc_score(y_true, y_prob)
            fpr, tpr, _ = roc_curve(y_true, y_prob)
            if plot_roc:
                if ax is None:
                    plt.figure()
                    plt.plot(fpr, tpr, label=f"{model_name} (AUC={auc:.4f})")
                    plt.plot([0,1],[0,1], 'k--')
                    plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(); plt.title(f"ROC - {model_name}")
                    plt.show()
                else:
                    ax.plot(fpr, tpr, label=f"{model_name} (AUC={auc:.4f})")
        except Exception:
            auc = None
    else:
        try:
            auc = roc_auc_score(y_true, y_pred)
        except Exception:
            auc = None
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    return {"accuracy": acc, "f1": f1, "auc": auc}

# ---------- Models ----------
class MLP(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, 128)
        self.bn1 = nn.BatchNorm1d(128)
        self.fc2 = nn.Linear(128, 64)
        self.bn2 = nn.BatchNorm1d(64)
        self.drop = nn.Dropout(0.3)
        self.fc3 = nn.Linear(64, 32)
        self.out = nn.Linear(32, 2)
    def forward(self, x):
        x = F.relu(self.bn1(self.fc1(x)))
        x = self.drop(F.relu(self.bn2(self.fc2(x))))
        x = F.relu(self.fc3(x))
        return self.out(x)

def train_pytorch(model, train_loader, epochs=20, lr=1e-3, device=DEVICE, name="Model"):
    model.to(device)
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.CrossEntropyLoss()
    losses = []
    for ep in range(epochs):
        model.train()
        running = 0.0
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            opt.zero_grad()
            out = model(xb)
            loss = loss_fn(out, yb)
            loss.backward()
            opt.step()
            running += loss.item() * xb.size(0)
        epoch_loss = running / len(train_loader.dataset)
        losses.append(epoch_loss)
        if (ep+1) % max(1, (epochs//4)) == 0:
            print(f"{name} Epoch {(ep+1)}/{epochs} loss: {epoch_loss:.5f}")
    return model, losses

# ---------- FT-Transformer (minimal, numerical-only, self-contained) ----------
class FeatureTokenizer(nn.Module):
    """
    Projects each numerical feature independently to d_token, then sums with [CLS] token.
    """
    def __init__(self, n_num_features, d_token):
        super().__init__()
        self.n = n_num_features
        self.d = d_token
        self.weight = nn.Parameter(torch.randn(n_num_features, d_token) * 0.02)
        self.bias = nn.Parameter(torch.zeros(n_num_features, d_token))
        self.cls = nn.Parameter(torch.randn(1, d_token) * 0.02)

    def forward(self, x):
        # x: (batch, n_features)
        # tokenize each feature: (batch, n, d)
        x = x.unsqueeze(-1) * self.weight.unsqueeze(0) + self.bias.unsqueeze(0)
        # prepend CLS: (batch, 1, d)
        cls = self.cls.unsqueeze(0).expand(x.size(0), -1, -1)
        return torch.cat([cls, x], dim=1)

class FTTransformer(nn.Module):
    def __init__(self, n_num_features, d_token=64, n_heads=8, n_layers=3, dropout=0.1, n_classes=2):
        super().__init__()
        self.tok = FeatureTokenizer(n_num_features, d_token)
        enc_layer = nn.TransformerEncoderLayer(d_model=d_token, nhead=n_heads, dim_feedforward=d_token*4, dropout=dropout, batch_first=True, activation="gelu", norm_first=True)
        self.encoder = nn.TransformerEncoder(enc_layer, num_layers=n_layers)
        self.norm = nn.LayerNorm(d_token)
        self.head = nn.Sequential(
            nn.Linear(d_token, d_token),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(d_token, n_classes)
        )

    def forward(self, x):
        # x: (batch, n_features)
        tok = self.tok(x)              # (batch, 1+n_features, d)
        enc = self.encoder(tok)        # (batch, 1+n_features, d)
        cls = self.norm(enc[:, 0, :])  # (batch, d)
        return self.head(cls)

# ---------- Cross-validation ----------
scaler_global = StandardScaler()
X_all = scaler_global.fit_transform(X_raw_all.values)

skf = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=RANDOM_SEED)

results_per_model = {
    "MLP": [],
    "TabNet": [],
    "FT-Transformer": []
}

roc_curves = {
    "MLP": [],
    "TabNet": [],
    "FT-Transformer": []
}

fold_idx = 0
for train_idx, test_idx in skf.split(X_all, y_all):
    fold_idx += 1
    print(f"\n========== Fold {fold_idx}/{N_SPLITS} ==========")
    X_train, X_test = X_all[train_idx], X_all[test_idx]
    y_train, y_test = y_all[train_idx], y_all[test_idx]

    # tensor datasets
    train_ds = TensorDataset(torch.tensor(X_train, dtype=torch.float32), torch.tensor(y_train, dtype=torch.long))
    test_ds  = TensorDataset(torch.tensor(X_test,  dtype=torch.float32), torch.tensor(y_test,  dtype=torch.long))
    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False)

    # ----- MLP -----
    mlp = MLP(X_train.shape[1])
    mlp, mlp_losses = train_pytorch(mlp, train_loader, epochs=MLP_EPOCHS, lr=LR, device=DEVICE, name=f"MLP (fold {fold_idx})")
    plt.figure(); plt.plot(mlp_losses); plt.title(f"MLP Training Loss (fold {fold_idx})"); plt.xlabel("Epoch"); plt.show()

    mlp.eval()
    probs_mlp=[]; preds_mlp=[]
    with torch.no_grad():
        for xb, _ in test_loader:
            xb = xb.to(DEVICE)
            out = mlp(xb)
            p = F.softmax(out, dim=1)[:,1].cpu().numpy()
            pr = out.argmax(dim=1).cpu().numpy()
            probs_mlp.extend(p); preds_mlp.extend(pr)
    mlp_res = evaluate_model(y_test, np.array(preds_mlp), np.array(probs_mlp), model_name=f"MLP (fold {fold_idx})")
    results_per_model["MLP"].append(mlp_res)
    try:
        fpr, tpr, _ = roc_curve(y_test, np.array(probs_mlp))
        roc_curves["MLP"].append((fpr, tpr, mlp_res["auc"]))
    except:
        pass

    # SHAP for MLP (GradientExplainer on small subset)
    print("\nComputing SHAP for MLP (GradientExplainer, subset)...")
    try:
        be_n = min(100, X_train.shape[0]); te_n = min(100, X_test.shape[0])
        background = torch.tensor(X_train[np.random.choice(X_train.shape[0], be_n, replace=False)], dtype=torch.float32).to(DEVICE)
        test_sample = torch.tensor(X_test[:te_n], dtype=torch.float32).to(DEVICE)
        explainer_mlp = shap.GradientExplainer(mlp, background)
        shap_vals = explainer_mlp.shap_values(test_sample)
        shap.summary_plot(shap_vals[1], X_test[:te_n], feature_names=features, show=True)
    except Exception as e:
        print("MLP SHAP failed:", e)

    # ----- TabNet -----
    print("\nTraining TabNet (may take a bit)...")
    tabnet = TabNetClassifier(seed=RANDOM_SEED, verbose=0)
    start = time.time()
    tabnet.fit(
        X_train, y_train,
        eval_set=[(X_test, y_test)],
        max_epochs=TABNET_MAX_EPOCHS,
        patience=15,
        batch_size=BATCH_SIZE,
        virtual_batch_size=32
    )
    print("TabNet runtime (s):", time.time()-start)

    y_pred_tab = tabnet.predict(X_test)
    y_prob_tab = tabnet.predict_proba(X_test)[:,1]
    tabnet_res = evaluate_model(y_test, y_pred_tab, y_prob_tab, model_name=f"TabNet (fold {fold_idx})")
    results_per_model["TabNet"].append(tabnet_res)
    try:
        fpr, tpr, _ = roc_curve(y_test, y_prob_tab)
        roc_curves["TabNet"].append((fpr, tpr, tabnet_res["auc"]))
    except:
        pass

    # TabNet SHAP (fallback to model-agnostic Explainer with small background)
    print("\nComputing SHAP for TabNet (small subset)...")
    try:
        be_n = min(200, X_train.shape[0]); te_n = min(100, X_test.shape[0])
        explainer_tab = shap.Explainer(tabnet.predict_proba, X_train[np.random.choice(X_train.shape[0], be_n, replace=False)])
        shap_tab = explainer_tab(X_test[:te_n])
        shap.summary_plot(shap_tab, X_test[:te_n], feature_names=features, show=True)
    except Exception as e:
        print("TabNet SHAP failed or too slow:", e)
        # Built-in importances as fallback plot
        try:
            imp = tabnet.feature_importances_
            imp_df = pd.DataFrame({"feature": features, "importance": imp}).sort_values("importance", ascending=False)
            plt.figure(figsize=(6,3))
            sns.barplot(x="importance", y="feature", data=imp_df)
            plt.title("TabNet feature importances")
            plt.show()
        except:
            pass

    # ----- FT-Transformer -----
    ftt = FTTransformer(n_num_features=X_train.shape[1], d_token=64, n_heads=8, n_layers=3, dropout=0.1, n_classes=2).to(DEVICE)
    ftt, ftt_losses = train_pytorch(ftt, train_loader, epochs=FTT_EPOCHS, lr=LR, device=DEVICE, name=f"FT-Transformer (fold {fold_idx})")
    plt.figure(); plt.plot(ftt_losses); plt.title(f"FT-Transformer Training Loss (fold {fold_idx})"); plt.xlabel("Epoch"); plt.show()

    ftt.eval()
    probs_ftt=[]; preds_ftt=[]
    with torch.no_grad():
        for xb, _ in test_loader:
            xb = xb.to(DEVICE)
            out = ftt(xb)
            p = F.softmax(out, dim=1)[:,1].cpu().numpy()
            pr = out.argmax(dim=1).cpu().numpy()
            probs_ftt.extend(p); preds_ftt.extend(pr)
    ftt_res = evaluate_model(y_test, np.array(preds_ftt), np.array(probs_ftt), model_name=f"FT-Transformer (fold {fold_idx})")
    results_per_model["FT-Transformer"].append(ftt_res)
    try:
        fpr, tpr, _ = roc_curve(y_test, np.array(probs_ftt))
        roc_curves["FT-Transformer"].append((fpr, tpr, ftt_res["auc"]))
    except:
        pass

    # SHAP for FT-Transformer (GradientExplainer on small subset)
    print("\nComputing SHAP for FT-Transformer (GradientExplainer, subset)...")
    try:
        be_n = min(100, X_train.shape[0]); te_n = min(100, X_test.shape[0])
        background = torch.tensor(X_train[np.random.choice(X_train.shape[0], be_n, replace=False)], dtype=torch.float32).to(DEVICE)
        test_sample = torch.tensor(X_test[:te_n], dtype=torch.float32).to(DEVICE)
        explainer_ftt = shap.GradientExplainer(ftt, background)
        shap_vals_ftt = explainer_ftt.shap_values(test_sample)
        shap.summary_plot(shap_vals_ftt[1], X_test[:te_n], feature_names=features, show=True)
    except Exception as e:
        print("FT-Transformer SHAP failed:", e)

# ---------- Aggregate CV results ----------
def aggregate_results(res_list):
    df = pd.DataFrame(res_list)
    return df.mean(numeric_only=True).to_dict(), df

agg_summary = {}
detailed_frames = {}
for model_name, res_list in results_per_model.items():
    mean_metrics, df_detail = aggregate_results(res_list)
    agg_summary[model_name] = mean_metrics
    detailed_frames[model_name] = df_detail

print("\n===== Cross-Validation Summary (mean over folds) =====")
summary_df = pd.DataFrame(agg_summary).T
display(summary_df)

# ---------- Combined ROC curve per model (mean curve shown per fold set) ----------
plt.figure(figsize=(6,5))
for model_name, curves in roc_curves.items():
    # plot each fold curve lightly and annotate mean AUC
    aucs = []
    for fpr, tpr, auc in curves:
        aucs.append(auc if auc is not None else np.nan)
        plt.plot(fpr, tpr, alpha=0.3, label=None, color=None)
    if len(aucs) > 0 and not np.all(np.isnan(aucs)):
        label = f"{model_name} (mean AUC={np.nanmean(aucs):.4f})"
    else:
        label = f"{model_name}"
    # To avoid duplicate legends for multiple folds, just add a dummy line for the label
    plt.plot([0,1],[0,1], alpha=0)  # invisible trick to reserve space
    plt.plot([], [], label=label)
plt.plot([0,1],[0,1], 'k--', label="Random")
plt.xlabel("FPR"); plt.ylabel("TPR")
plt.title("ROC Curves across folds")
plt.legend()
plt.show()

# ---------- Choose best by AUC and retrain on full dataset ----------
best_model_name = None
if summary_df["auc"].notnull().any():
    best_model_name = summary_df["auc"].idxmax()
else:
    best_model_name = summary_df["f1"].idxmax()

print(f"\n✅ Best model by CV primary metric: {best_model_name}")

# Save scaler
import joblib
joblib.dump(scaler_global, os.path.join(ARTIFACT_DIR, "scaler.pkl"))

# Retrain the best model on full data and save
if best_model_name == "MLP":
    full_ds = TensorDataset(torch.tensor(X_all, dtype=torch.float32), torch.tensor(y_all, dtype=torch.long))
    full_loader = DataLoader(full_ds, batch_size=BATCH_SIZE, shuffle=True)
    model = MLP(X_all.shape[1]).to(DEVICE)
    model, _ = train_pytorch(model, full_loader, epochs=MLP_EPOCHS, lr=LR, device=DEVICE, name="MLP (full)")
    torch.save(model.state_dict(), os.path.join(ARTIFACT_DIR, "mlp_full.pth"))

elif best_model_name == "FT-Transformer":
    full_ds = TensorDataset(torch.tensor(X_all, dtype=torch.float32), torch.tensor(y_all, dtype=torch.long))
    full_loader = DataLoader(full_ds, batch_size=BATCH_SIZE, shuffle=True)
    model = FTTransformer(n_num_features=X_all.shape[1], d_token=64, n_heads=8, n_layers=3, dropout=0.1, n_classes=2).to(DEVICE)
    model, _ = train_pytorch(model, full_loader, epochs=FTT_EPOCHS, lr=LR, device=DEVICE, name="FT-Transformer (full)")
    torch.save(model.state_dict(), os.path.join(ARTIFACT_DIR, "ft_transformer_full.pth"))

else:  # TabNet
    model = TabNetClassifier(seed=RANDOM_SEED, verbose=0)
    model.fit(
        X_all, y_all,
        max_epochs=TABNET_MAX_EPOCHS,
        patience=15,
        batch_size=BATCH_SIZE,
        virtual_batch_size=32
    )
    try:
        model.save_model(os.path.join(ARTIFACT_DIR, "tabnet_full.zip"))
    except Exception as e:
        print("TabNet save failed:", e)

print("Artifacts saved to", ARTIFACT_DIR)

print("\nTIPS:")
print("- SHAP calls use small subsets to avoid long runtimes. Increase only if you have GPU/time.")
print("- Replace crop thresholds with authoritative agronomy values for real recommendations (if you bring that section back).")
print("- The CV summary gives mean AUC/F1/Accuracy across folds. Consider reporting std as well for robustness.")
```
